```python
# ==========================================
#  PROYECTO 1: EL LABERINTO (NIVEL EXPLORADOR)
# ==========================================
# Objetivo: Llevar al rat贸n () hasta el queso ()
# Conceptos: Listas, Coordenadas (x,y), Ciclo While, Condicionales If/Else

```

```python
# --- 1. CONFIGURACIN VISUAL ---
# Definimos las "piezas" de nuestro tablero usando emojis.
# Esto hace que el c贸digo sea f谩cil de leer y modificar.
PARED = "猬"   # Representa un muro
CAMINO = "猬"  # Representa espacio libre
JUGADOR = "" # Tu personaje
META = ""    # El objetivo

```

```python
# --- 2. DEFINICIN DEL MAPA ---
# Usamos una "Lista de Listas" (Matriz) para dibujar el nivel.
# Puedes cambiar el dise帽o moviendo los cuadrados aqu铆.
mapa = [
    [JUGADOR, CAMINO, PARED, CAMINO, CAMINO],
    [PARED, CAMINO, PARED, CAMINO, PARED],
    [CAMINO, CAMINO, CAMINO, CAMINO, PARED],
    [CAMINO, PARED, PARED, CAMINO, CAMINO],
    [CAMINO, CAMINO, CAMINO, PARED, META]
]

```

```python
# Posici贸n inicial del jugador (Fila 0, Columna 0)
# Python empieza a contar desde 0, no desde 1.
fila_jugador = 0
col_jugador = 0

print("隆Bienvenido al Laberinto IA!")
print("Instrucciones: Escribe w (arriba), s (abajo), a (izquierda), d (derecha)")
print("Escribe 'salir' para rendirte.\n")

```

```python
# --- 3. CICLO PRINCIPAL DEL JUEGO ---
# 'While True' significa que esto se repetir谩 infinitamente
# hasta que usemos la instrucci贸n 'break'.
juego_activo = True

while juego_activo:
    
    # A. DIBUJAR EL MAPA EN PANTALLA
    print("\n--- MAPA ACTUAL ---")
    for fila in mapa:
        # .join une los elementos de la lista en una sola l铆nea de texto
        print("".join(fila))
    print("-------------------")

    # B. VERIFICAR SI GANAMOS
    # Preguntamos: 驴En mi posici贸n actual, estoy pisando el queso?
    if mapa[fila_jugador][col_jugador] == META:
        print("\n 隆FELICIDADES! 隆El rat贸n se comi贸 el queso! ")
        print("Has completado el Nivel 1.")
        juego_activo = False # Apagamos el interruptor del ciclo
        break 

    # C. PEDIR MOVIMIENTO AL JUGADOR
    movimiento = input("驴Hacia d贸nde te mueves? (w/a/s/d): ").lower()

    # Guardamos la posici贸n actual en variables temporales "futuras"
    # para probar si el movimiento es v谩lido ANTES de hacerlo real.
    fila_futura = fila_jugador
    col_futura = col_jugador

    # D. CALCULAR INTENCIN DE MOVIMIENTO
    if movimiento == "w": fila_futura -= 1    # Arriba (restar fila)
    elif movimiento == "s": fila_futura += 1  # Abajo (sumar fila)
    elif movimiento == "a": col_futura -= 1   # Izquierda (restar columna)
    elif movimiento == "d": col_futura += 1   # Derecha (sumar columna)
    elif movimiento == "salir":
        print("Juego terminado.")
        break
    else:
        print("Tecla no v谩lida. Intenta de nuevo.")
        continue # 'Continue' salta al inicio del ciclo sin hacer nada m谩s

    # E. VALIDAR EL MOVIMIENTO (L贸gica del Juego)
    
    # Regla 1: No salir del mapa (L铆mites)
    # El mapa mide 5x5, as铆 que los 铆ndices v谩lidos son 0, 1, 2, 3, 4.
    if fila_futura < 0 or fila_futura > 4 or col_futura < 0 or col_futura > 4:
        print("锔 隆Cuidado! Te vas a caer del mapa.")
        continue
    
    # Regla 2: No atravesar paredes
    # Miramos en el mapa qu茅 hay en la casilla futura
    if mapa[fila_futura][col_futura] == PARED:
        print("Chocaste contra una pared.")
        continue

    # F. ACTUALIZAR EL MAPA (Si pas贸 todas las reglas)
    
    # 1. Borramos al rat贸n de su posici贸n vieja (ponemos suelo blanco)
    if mapa[fila_jugador][col_jugador] != META: 
         mapa[fila_jugador][col_jugador] = CAMINO
    
    # 2. Actualizamos las coordenadas oficiales
    fila_jugador = fila_futura
    col_jugador = col_futura
    
    # 3. Dibujamos al rat贸n en la nueva posici贸n
    # (Solo si no es la meta, para no borrar el queso visualmente antes de ganar)
    if mapa[fila_jugador][col_jugador] != META:
        mapa[fila_jugador][col_jugador] = JUGADOR
```
